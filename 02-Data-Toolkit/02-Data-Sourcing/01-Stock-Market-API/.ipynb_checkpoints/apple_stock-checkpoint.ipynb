# Stock Market Analysis

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
```

## Apple last 3 months

Last 3 months of Apple stock prices

```python
url_apple = "https://iex.lewagon.com/stable/stock/aapl/chart/3m"
response_apple = requests.get(url_apple).json()
```

```python
apple_df = pd.DataFrame(response_apple)
```

```python
# function to change date type and set index to date

def change_date(df):
    df.date = pd.to_datetime(df.date, format="%Y-%m-%d")
    df.set_index("date", inplace=True)
    return df
```

```python
apple_df = change_date(apple_df)
```

```python
apple_df.close.plot();
```

```python
apple_df[["close", "open", "low", "high"]].plot(figsize=(16,6));
```

## GAFA stocks

Google, Apple, Facebook and Amazon Stocks

```python
url_google = "https://iex.lewagon.com/stable/stock/googl/chart/3m"
url_amazon = "https://iex.lewagon.com/stable/stock/amzn/chart/3m"
url_facebook = "https://iex.lewagon.com/stable/stock/fb/chart/3m"
```

```python
response_google = requests.get(url_google).json()
response_amazon = requests.get(url_amazon).json()
response_facebook = requests.get(url_facebook).json()
```

```python
google_df = pd.DataFrame(response_google)
amazon_df = pd.DataFrame(response_amazon)
facebook_df = pd.DataFrame(response_facebook)
```

```python
google_df = change_date(google_df)
amazon_df = change_date(amazon_df)
facebook_df = change_date(facebook_df)
```

```python
merge_df = pd.merge(left=google_df["close"], right=amazon_df["close"], left_index=True, right_index=True)
```

```python
merge_df.rename(columns={"close_x": "Google", "close_y": "Amazon"}, inplace=True)
```

```python
merge_df = pd.merge(left=merge_df, right=facebook_df["close"], left_index=True, right_index=True)
```

```python
merge_df = pd.merge(left=merge_df, right=apple_df["close"], left_index=True, right_index=True)
```

```python
merge_df.rename(columns={"close_x": "Facebook", "close_y": "Apple"}, inplace=True)
```

```python
merge_df.plot(figsize=(16,8));
```

###  Normalization

```python
def normalize_data(data):
    return (data - np.min(data)) / (np.max(data) - np.min(data))
```

```python
merge_df.apply(normalize_data).plot(figsize=(16,8));
```
